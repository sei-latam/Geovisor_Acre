# Import libraries

```python
import os
import re
```

```python
base_path = r"C:\Users\Carlos Andres\Desktop\Geovisor_Rio_Acre\Output_HEC_RAS_Rasters"
carpetas_tr = ["TR2", "TR10", "TR50", "TR100", "TR2023"]

patron = re.compile(r"Depth\s*\((.*?)\).*?\.tif", re.IGNORECASE)

for carpeta in carpetas_tr:
    path_completo = os.path.join(base_path, carpeta)
    if not os.path.exists(path_completo):
        continue

    numero_tr = re.findall(r"\d+", carpeta)[0]
    tr_formateado = f"TR{int(numero_tr):02d}"
    
    archivos = [f for f in os.listdir(path_completo) if f.lower().endswith('.tif') and patron.search(f)]
    archivos.sort()
    
    for idx, archivo_orig in enumerate(archivos, start=1):
        coincidencia = patron.search(archivo_orig)
        if coincidencia:
            fecha_hora_sucia = coincidencia.group(1)
            fecha_hora_limpia = fecha_hora_sucia.strip().replace(" ", "_")
            step = f"{idx:02d}"
            nuevo_nombre = f"Depth_{step}_{fecha_hora_limpia}_{tr_formateado}.tif"
            ruta_antigua = os.path.join(path_completo, archivo_orig)
            ruta_nueva = os.path.join(path_completo, nuevo_nombre)
            os.rename(ruta_antigua, ruta_nueva)
```
