### HDF5 Structure of file ACRE_HIDRODIN.p04

``````````
🗂️ Raíz del Archivo HDF5 (Plan de HEC-RAS)
│
├── 📂 Event Conditions/                                         <-- Configuración hidrológica de entrada
│   └── 📂 Boundary Conditions/                                  <-- Datos de las condiciones de contorno
│       ├── 📂 Stage Hydrographs/                                <-- Hidrogramas de nivel/etapa
│       ├── 📂 Flow Hydrographs/                                 <-- Hidrogramas de caudal de entrada (Arroyos/Ríos)
│       └── 📄 Boundary Condition Connections                    <-- Vinculación de hidrogramas con la malla 2D
│
├── 📂 Geometry/                                                 <-- Toda la infraestructura geográfica y física
│   ├── 📄 Land Cover (Manning's n)                              <-- Capas de uso de suelo y coeficientes de rugosidad
│   ├── 📂 2D Flow Areas/                                        <-- El corazón geométrico bidimensional
│   │   └── 📂 2d_COBIJA/                                  
│   │       ├── 📄 Attributes                                    <-- Metadatos de la malla (Número de celdas, caras, etc.)
│   │       ├── 📄 Cells Center Coordinate                       <-- Coordenadas (X, Y) de los centros de las celdas
│   │       ├── 📄 Cells Face Courant Info                       <-- Datos geométricos internos para calcular el No. Courant
│   │       ├── 📄 Cells Minimum Elevation                       <-- Elevación de fondo (Z) de cada una de las celdas
│   │       ├── 📄 FacePoints Coordinate                         <-- Coordenadas (X, Y) de los vértices de las esquinas
│   │       ├── 📄 Faces Cell Indexes                            <-- Conectividad: Qué celdas están a la izq/der de cada cara
│   │       ├── 📄 Faces FacePoint Indexes                       <-- Conectividad: Qué vértices forman los extremos de cada cara
│   │       ├── 📄 Faces Low Elevation Centroid                  <-- Elevación (Z) del punto más bajo de cada cara
│   │       ├── 📄 Faces Normal Unit Vector                      <-- Vectores de dirección hidráulica del flujo en las caras
│   │       └── 📄 Faces Perimeter                               <-- Longitudes físicas de las caras de la malla
│   └── 📂 Structures/                                           <-- Puentes, alcantarillas o diques incorporados (si aplican)
│
├── 📂 Results/                                                  <-- El almacén de outputs del modelo numérico
│   └── 📂 Unsteady/
│       ├── 📂 Log Output/                                       <-- Diagnóstico y bitácora de la simulación
│       │   ├── 📄 Computations Log                              <-- Registro de la CPU (Tiempo de corrida, procesadores)
│       │   └── 📄 Summary of Errors                             <-- Lista de errores de convergencia y advertencias hidráulicas
│       │
│       └── 📂 Output/
│           └── 📂 Output Blocks/
│               ├── 📂 Base Output/                              <-- RESULTADOS HIDRÁULICOS PRINCIPALES
│               │   ├── 📂 2D Flow Areas/
│               │   │   └── 📂 2d_COBIJA/
│               │   │       └── 📄 Cell Minimum Elevation        <-- Elevaciones de control post-procesadas
│               │   │
│               │   ├── 📂 Unsteady Time Series/                 <-- Datos dinámicos a lo largo de las 325 simulaciones
│               │   │   ├── 📂 2D Flow Areas/
│               │   │   │   └── 📂 2d_COBIJA/
│               │   │   │       ├── 📄 Face Velocity             <-- Matriz dinámica de velocidad
│               │   │   │       └── 📄 Water Surface             <-- Matriz dinámica de niveles WSE
│               │   │   └── 📄 Time Date Stamp                   <-- El calendario con las fechas reales
│               │   │
│               │   └── 📂 Summary Output/                       <-- Los valores pico/extremos de toda la modelación
│               │       └── 📂 2D Flow Areas/
│               │           └── 📂 2d_COBIJA/
│               │               ├── 📄 Maximum Face Velocity     <-- Mapa de velocidad máxima en caras
│               │               └── 📄 Maximum Water Surface     <-- Mapa de nivel máximo (Mancha de inundación)
│               │
│               └── 📂 Computations/                             <-- Parámetros numéricos internos de la simulación
│                   ├── 📂 2D Flow Areas/
│                   │   └── 📂 2d_COBIJA/
│                   │       ├── 📄 Cell Last Iteration           <-- Cuántas iteraciones le tomó estabilizar cada celda
│                   │       ├── 📄 Maximum Face Courant          <-- El número de Courant máximo por cara
│                   │       ├── 📄 Maximum Water Surface Error   <-- Errores numéricos de balance de masa
│                   │       └── 📄 Starting Differences WSE       <-- Diferencia de nivel respecto al estado inicial
│                   └── 📄 Time Step Iterations                  <-- Historial de iteraciones por cada paso de tiempo
│
└── 📂 Plan Data/                                                <-- Configuración y opciones del Plan de HEC-RAS
    ├── 📄 Plan Information                                      <-- Nombre del Plan, descripción, versión de HEC-RAS usada
    ├── 📄 Plan Parameters                                       <-- Ecuaciones usadas (Difusiva vs. Saint-Venant Completa)
    └── 📄 Time Window                                           <-- Fecha de inicio, fecha de fin y Time Step
``````````

##### Export maximum water surface summary to GIS Data (vector, raster and geojson)

> [Sum_WSEL] (Este nombre representa los valores WSEL del apartado summary calculado por HEC-RAS que se van a exportar tanto al shapefile como al raster.)

> ['Sum_WSEL': datos_wse_max[0, :]] (convierte a "Sum_WSEL" los resultados WSEL del apartado summary para que se exporten con ese nombre tanto al shapefile como al raster.)

```````````python
out_GIS_vector_WSE_sum = os.path.join(output_GIS_vector, "Max_WSEL_Summary")
os.makedirs(out_GIS_vector_WSE_sum, exist_ok=True)
out_GIS_raster_WSE_sum = os.path.join(output_GIS_raster, "Max_WSEL_Summary")
os.makedirs(out_GIS_raster_WSE_sum, exist_ok=True)
out_GIS_geojson_WSE_sum = os.path.join(output_GIS_geojson, "Max_WSEL_Summary")
os.makedirs(out_GIS_geojson_WSE_sum, exist_ok=True)
campo_valor = "Sum_WSEL"

try:
    if not os.path.exists(plan_hdf_path):
        raise FileNotFoundError(f"Archivo no encontrado: {plan_hdf_path}")        
    nombre_archivo_completo = os.path.basename(plan_hdf_path)
    partes_nombre = nombre_archivo_completo.split('.')
    sufijo_plan = f"_{partes_nombre[-2]}" if len(partes_nombre) >= 2 else ""

    with h5py.File(plan_hdf_path, 'r') as hdf:
        malla_path = f"Geometry/2D Flow Areas/{nombre_malla}"        
        datos_wse_max = hdf[f"{RUTA_SUMMARY}/Maximum Water Surface"][:]
        wsel_valores = datos_wse_max[0, :]
        time_valores = datos_wse_max[1, :]
        
        posibles_nombres = ["Cells FacePoint Indexes", "Cells Facepoint Indexes", "Cell FacePoint Indexes"]
        cells_facepoint_idx = None
        for nombre in posibles_nombres:
            ruta_prueba = f"{malla_path}/{nombre}"
            if ruta_prueba in hdf:
                cells_facepoint_idx = hdf[ruta_prueba][:]
                break

        poligonos_celdas = []
        indices_validos = []

        if cells_facepoint_idx is None:
            facepoints_coords = hdf[f"{malla_path}/FacePoints Coordinate"][:]
            face_to_facepoints = hdf[f"{malla_path}/Faces FacePoint Indexes"][:]
            cells_face_idx = hdf[f"{malla_path}/Cells Face Indexes"][:]            
            total_celdas = cells_face_idx.shape[0]
            
            for idx_celda in range(total_celdas):
                if time_valores[idx_celda] <= 0:
                    continue
                    
                caras_idx = cells_face_idx[idx_celda]
                caras_idx = caras_idx[caras_idx >= 0]
                v_indices = set()
                for f_idx in caras_idx:
                    v_indices.update(face_to_facepoints[f_idx])
                v_indices.discard(-1)
                
                vertices_idx = np.array(list(v_indices))
                if len(vertices_idx) >= 3:
                    coords_poligono = facepoints_coords[vertices_idx]
                    cx, cy = coords_poligono[:, 0].mean(), coords_poligono[:, 1].mean()
                    angulos = np.arctan2(coords_poligono[:, 1] - cy, coords_poligono[:, 0] - cx)
                    coords_poligono = coords_poligono[np.argsort(angulos)]
                    poligonos_celdas.append(Polygon(coords_poligono))
                    indices_validos.append(idx_celda)
                    
        else:
            facepoints_coords = hdf[f"{malla_path}/FacePoints Coordinate"][:]            
            for idx_celda in range(len(wsel_valores)):
                if time_valores[idx_celda] <= 0:
                    continue
                    
                vertices_idx = cells_facepoint_idx[idx_celda]
                vertices_idx = vertices_idx[vertices_idx >= 0]
                
                if len(vertices_idx) >= 3:
                    coords_poligono = facepoints_coords[vertices_idx]
                    poligonos_celdas.append(Polygon(coords_poligono))
                    indices_validos.append(idx_celda)

        df_export = pd.DataFrame({'Sum_WSEL': wsel_valores[indices_validos],'Time': time_valores[indices_validos]})
        gdf_gis = gpd.GeoDataFrame(df_export, geometry=poligonos_celdas, crs=crs_proyecto)
        nombre_base_archivos = f"Summary_Max_Water_Surface_Elevation{sufijo_plan}"
        ruta_shapefile = os.path.join(out_GIS_vector_WSE_sum, f"{nombre_base_archivos}.shp")
        gdf_gis.to_file(ruta_shapefile, driver='ESRI Shapefile') 
        print(f"--------> Archivo vectorial creado en: {os.path.abspath(ruta_shapefile)}")

    if not os.path.exists(ruta_shapefile):
        raise FileNotFoundError(f"Archivo vectorial no encontrado: {ruta_shapefile}")
    
    gdf_para_raster = gpd.read_file(ruta_shapefile)
    if gdf_para_raster.crs is None:
        gdf_para_raster.set_crs(crs_proyecto, inplace=True)
    ruta_salida_raster = os.path.join(out_GIS_raster_WSE_sum, f"{nombre_base_archivos}.tif")
    cube = make_geocube(vector_data=gdf_para_raster, measurements=[campo_valor], resolution=(-tamano_celda, tamano_celda), fill=np.nan)
    raster_final = cube[campo_valor]
    raster_final.rio.to_raster(ruta_salida_raster, driver="GTiff", compress="LZW")
    print(f"--------> Archivo raster creado en: {os.path.abspath(ruta_salida_raster)}")

    ruta_salida_geojson = os.path.join(out_GIS_geojson_WSE_sum, f"{nombre_base_archivos}.geojson")
    gdf_para_raster.to_file(ruta_salida_geojson, driver='GeoJSON')
    print(f"--------> Archivo GeoJSON creado en: {os.path.abspath(ruta_salida_geojson)}")

except Exception as e:
    print(f"Error en el procesamiento: {e}")
```````````

```````````python
out_GIS_vector_VEL_series = os.path.join(output_GIS_vector, "VEL_Time_Series")
os.makedirs(out_GIS_vector_VEL_series, exist_ok=True)
out_GIS_raster_VEL_series = os.path.join(output_GIS_raster, "VEL_Time_Series")
os.makedirs(out_GIS_raster_VEL_series, exist_ok=True)
out_GIS_geojson_VEL_series = os.path.join(output_GIS_geojson, "VEL_Time_Series")
os.makedirs(out_GIS_geojson_VEL_series, exist_ok=True)
campo_valor = "TS_VEL"

try:
    if not os.path.exists(plan_hdf_path):
        raise FileNotFoundError(f"Archivo no encontrado: {plan_hdf_path}")
        
    nombre_archivo_completo = os.path.basename(plan_hdf_path)
    partes_nombre = nombre_archivo_completo.split('.')
    sufijo_plan = f"_{partes_nombre[-2]}" if len(partes_nombre) >= 2 else ""

    with h5py.File(plan_hdf_path, 'r') as hdf:
        if RUTA_TIME_STAMP in hdf:
            timestamps_bytes = hdf[RUTA_TIME_STAMP][:]
            lista_fechas = [t.decode('utf-8').strip() if isinstance(t, bytes) else str(t).strip() for t in timestamps_bytes]
        else:
            lista_fechas = []
            
        facepoint_coords = hdf[f"Geometry/2D Flow Areas/{nombre_malla}/FacePoints Coordinate"][:]
        face_to_facepoints = hdf[f"Geometry/2D Flow Areas/{nombre_malla}/Faces FacePoint Indexes"][:]
        vel_matriz = hdf[f"{RUTA_SERIES}/Face Velocity"][:]        
        
        p1_coords = facepoint_coords[face_to_facepoints[:, 0]]
        p2_coords = facepoint_coords[face_to_facepoints[:, 1]]
        coords_caras = (p1_coords + p2_coords) / 2.0
        
        total_pasos = vel_matriz.shape[0]
        if len(lista_fechas) < total_pasos:
            lista_fechas = [f"Paso #{i+1}" for i in range(total_pasos)]
            
        for paso in range(0, total_pasos, intervalo_muestreo):
            fecha_hora_real = lista_fechas[paso]
            valores_paso = vel_matriz[paso, :]
            
            df_paso = pd.DataFrame({'X': coords_caras[:, 0], 'Y': coords_caras[:, 1], 'TS_VEL': valores_paso})
            df_filtrado = df_paso.loc[df_paso['TS_VEL'] > 0.001].copy()
            if df_filtrado.empty:
                print(f"[{paso+1}/{total_pasos}] {fecha_hora_real}: Sin caras con flujo en este paso. Saltando.")
                continue
                
            geometria_puntos = [Point(xyz) for xyz in zip(df_filtrado['X'], df_filtrado['Y'])]
            gdf_paso = gpd.GeoDataFrame(df_filtrado, geometry=geometria_puntos, crs=crs_proyecto)            
            gdf_gis = gdf_paso[['TS_VEL', 'geometry']]
            
            fecha_limpia = re.sub(r'[\s:]+', '_', fecha_hora_real)
            nombre_base_archivos = f"VEL_{paso+1:02d}_{fecha_limpia}{sufijo_plan}"
            
            ruta_shapefile = os.path.join(out_GIS_vector_VEL_series, f"{nombre_base_archivos}.shp")
            gdf_gis.to_file(ruta_shapefile, driver='ESRI Shapefile')
            print(f"[{paso+1}/{total_pasos}] Vector guardado: {nombre_base_archivos}.shp")
            
            if not os.path.exists(ruta_shapefile):
                raise FileNotFoundError(f"Archivo vectorial no encontrado para rasterizar: {ruta_shapefile}")
            
            gdf_para_raster = gpd.read_file(ruta_shapefile)
            if gdf_para_raster.crs is None:
                gdf_para_raster.set_crs(crs_proyecto, inplace=True)
                
            ruta_salida_raster = os.path.join(out_GIS_raster_VEL_series, f"{nombre_base_archivos}.tif")
            
            cube = make_geocube(vector_data=gdf_para_raster,measurements=[campo_valor],resolution=(-tamano_celda, tamano_celda),fill=np.nan)
            raster_final = cube[campo_valor]
            raster_final.rio.to_raster(ruta_salida_raster,driver="GTiff",compress="LZW")
            print(f"[{paso+1}/{total_pasos}] Raster guardado: {nombre_base_archivos}.tif")

            ruta_salida_geojson = os.path.join(out_GIS_geojson_VEL_series, f"{nombre_base_archivos}.geojson")
            gdf_para_raster.to_file(ruta_salida_geojson, driver='GeoJSON')
            print(f"[{paso+1}/{total_pasos}] GeoJSON guardado: {nombre_base_archivos}.geojson")

except Exception as e:
    print(f"\nError al ejecutar la serie de tiempo: {e}")
```````````
